```python
import geopandas as gpd 
import pandas as pd
import numpy as np
import os
from shapely.geometry import MultiPoint, Point
```

```python
folder = "/media/riannek/minimax/gleis/gso"
files = [
"5556_km8,5-30,5+5560_km27,2-29,9_Run14.csv",
"5560_km27-33.2+5556_km8.6-33_Run24_ohne_Flughafentunnel.csv",
]



file = files[1]

df = pd.read_csv(os.path.join(folder, file), header=1, sep=";", decimal=",")
df.head()
```

```python
df['geometry'] = df.apply(
    lambda row: MultiPoint([
        (row['EastingL[m]'], row['NorthingL[m]']),
        (row['EastingC[m]'], row['NorthingC[m]']),
        (row['EastingR[m]'], row['NorthingR[m]'])
    ]),
    axis=1
)
gdf = gpd.GeoDataFrame(df, geometry='geometry', crs="EPSG:5684")
```

```python
gdf.head()
```

```python
gdf.to_file(os.path.join(folder, "gso.gpkg"), layer=file, driver="GPKG")
```
